Combining in pandas

```python
import pandas as pd
```

```python
canadian_videos = pd.read_csv('C:/Users/user/Desktop/data-science-journey/Datasets/CAvideos.csv')
british_videos = pd.read_csv('C:/Users/user/Desktop/data-science-journey/Datasets/GBvideos.csv')

pd.concat([canadian_videos, british_videos]).head()
```

```python
left = canadian_videos.set_index(['title', 'trending_date'])
right = british_videos.set_index(['title', 'trending_date'])


left.join(right, lsuffix = '_CAN', rsuffix = '_UK').head(4)
```
