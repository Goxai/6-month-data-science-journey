```python
# factors causing customer churn at the telco company


tel.groupby('Gender')['Churn'].value_counts().unstack()
```

```python
tel['Churn'].value_counts()
```

```python
# with this we will be able to get the mean of the churn of every factor we will look into

tel['churn_num'] = tel['Churn'].map({'No' : 0, 'Yes' : 1})
tel.head()
```

```python
# gender causes churn?

tel.groupby('Gender')['churn_num'].mean()

# from this it is obvious that gender plays a very minimal in customer churn as females has ~26.9% churn rate a little above 
# male with ~26.1%.
```

```python
# senior citizens causes churn?

tel.groupby('SeniorCitizen')['churn_num'].mean()

# we can see that the senior citizens churn rate of ~41.6% is almost twice of the non senior citizens.
# which means that the older you get, the tendencies of continuously using telcos reduces.
# so senior citizen increase churn.
```

```python
# internet service affect churn?

tel.groupby('InternetService')['churn_num'].mean()

# it is very clear that the internet service plays a major role in customer churn,
# as fiber optic has a churn rate of ~41.8% which is more twice the DSL churnrate of ~18.9%
# and 5x more than No churn rate of ~7.4%.
# what this means is that internet service play a very major role in customer churn at the telcos.
```

```python
tel['InternetService'].unique()
```

```python
# tech support causes churn?

tel.groupby('TechSupport')['churn_num'].mean()


# from the data, tech support is another key factor increasing customer churn.
# from the data, no tech support has ~41.6% churn rate which is more than twice when tech support is available and more than
# 5x when no internet service.
# tech support is another factor that play a major role in customer churn.
```

```python
tel['TechSupport'].unique()
```

```python
# monthly chrge causes churn?

tel.groupby(pd.cut(tel['MonthlyCharges'], bins = [0.0, 40.0, 70.0, 119.0], labels = ['low', 'mid', 'high']))['churn_num'].mean()

# from this we can see that monthly charge is another key factor causing customer churn,
# monthly charge of high monthly charge has a churn rate of ~35.3% which is thrice the churn rate of low monthly charge churn rate with ~11.6%
# also compared to mid charge with churn rate of ~23.9%.
```

```python
# to be able to convert TotalCharges from object to float 
# we fill the '' space first.

tel['TotalCharges'] = pd.to_numeric(tel['TotalCharges'], errors = 'coerce')

#tel['TotalCharges'].unique()
```

```python
tel['TotalCharges'].dtypes
```

```python
tel['TotalCharges_range'] = pd.cut(
    tel['TotalCharges'],
    bins = [0, 1500, 3000, 9000],
    labels = ['low', 'mid', 'high']
)
```

```python
tel.head()
```

```python
tel.groupby('TotalCharges_range')['churn_num'].mean()

# Customers with low total charges churn more because they are mostly new customers with short tenure.

# to further buttress my point, let check the tenure.
```

```python
tel.groupby(pd.cut(tel['Tenure'], bins = [0, 12, 36, 72]))['churn_num'].mean()

# You will see:
# 0â€“12 months â†’ very high churn
# 12â€“36 months â†’ medium churn
# 36â€“72 months â†’ low churn
#This proves your churn problem is mainly early customer dissatisfaction, not price alone.
```
