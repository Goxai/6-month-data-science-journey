```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'name': ['Alice', 'Bob', 'John', 'Mary'],
    'age': [20, 22, 19, 21],
    'score': [88, 95, 76, 90],
    'status': ['pass', 'pass', 'fail', 'pass'],
    'level': ['normal', 'high', 'normal', 'high']
}


df = pd.DataFrame(data)
print(df)
```

```python
# Data Exploration

df['name']
df[['name', 'score']]
df[df['status'] == 'pass']
```

```python
# line plot data visualisation

plt.plot(df['name'], df['score'])
plt.xlabel('name')
plt.ylabel('score')
plt.title('student performance ananlysis')
plt.show()
```

```python
# Bar chart data visualisation

plt.bar(df['name'], df['score'])
plt.xlabel('name')
plt.ylabel('score')
plt.title('student performance ananlysis')
plt.show()
```

```python
# histogram data visualisation

plt.hist(df['age'])
plt.xlabel('age')
plt.ylabel('age distribution')
plt.title('student performance ananlysis')
plt.show()
```
