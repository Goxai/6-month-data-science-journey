```python
sales = pd.read_csv('C:/Users/user/Desktop/data-science-journey/jupyter notebooks files/sales.csv')
print(sales)
```

```python
# calculate the total sales across different products and regions

total_sales = sales.sum()
print(total_sales)
```

```python
# calculate sales based on products

sales.sales.mean()
```

```python
# to claculate the sales based on prodcuts

product_sales_avg = sales.groupby('product')['sales'].sum()
print(product_sales_avg)
```

```python
# to claculate the sales based on location

region_sales_avg = sales.groupby('region')['sales'].sum()
print(region_sales_avg)
```

```python
# to sort the sales based on region

sorted_region = region_sales_avg.sort_values(ascending = False)
print(sorted_region)
```
