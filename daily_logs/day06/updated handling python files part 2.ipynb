```python
import csv

students = []

with open('students.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        row['age'] = int(row['age'])
        row['score'] = int(row['score'])
        students.append(row)
```

```python
#to get the highest score

top_student = max(students, key = lambda x : x['score'])
print(top_student)
```

```python
import csv

#find class statistics

scores = []
math_scores = []
eng_scores = []

with open('C:/Users/user/Desktop/data/students.csv', 'r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        math = int(row['Math'])
        eng = int(row['English'])

        scores.extend([math, eng])
        math_scores.append(math)
        eng_scores.append(eng)

print('class average:', sum(scores) / len(scores)),
print('math average:', sum(math_scores) / len(math_scores)),
print('eng average:', sum(eng_scores) / len(eng_scores))
```

```python
import csv

students = []

#read the available students csv file 

with open('C:/Users/user/Desktop/data/students.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        name = row[0]
        score1 = int(row[1])
        score2 = int(row[2])

         #calculate the average scores
        average = (score1 + score2) / 2

        #add reamrks
        
        remarks = 'Pass' if average >= 50 else 'fail'

        #update the row
        
        students.append((name, score1, score2, average, remarks))

#create a new =students file called staudents with remarks

with open('students_with_remarks.csv', 'w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'score1', 'score2', 'average', 'remarks'])
    writer.writerows(students)
print('new file created: students_with_remarks')
```

```python


#looking for a specific name on a file in python

import csv
def find_student(Name, filename = 'students_with_remarks.csv'):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row['Name'].strip().lower() == Name.strip().lower():
                return row
    return None

#using the function created above

student_file = find_student('Alice')
if student_file:
    print('found :', student_file)
else:
    print('student file not found')
```

```python
# adding student details to a file

import csv

def add_student(Name, score1, score2, filename = 'students_with_remarks.csv'):
    average = (score1 + score2) / 2
    remarks = 'pass' if average >= 50 else 'fail'

    with open(filename, 'a', newline = '') as file:
        writer = csv.writer(file)
        writer.writerow([Name, score1, score2, average, remarks])
    print(f'added {Name} (avg ={average}), remarks = {remarks}. ')
    
```

```python
# revision on how to add info to a file

import csv

def add_students_info(Name, score1, score2, filename = 'students_with_remarks.csv'):
    average = (score1 + score2) / 2
    remarks = 'Pass' if average >= 50 else 'Fail'

    with open(filename, 'a', newline = '') as file:
        writer = csv.writer(file)
        writer.writerow([Name, score1, score2, average, remarks])
        #writer = writerows(students)
    print(f'added {Name} (average = {average}, remarks = {remarks})')


#Example

new_students = add_students_info("Chris", 65, 72)
print(new_students)
```

```python

#deleting names from a file in file handling

import csv



def delete_student(Name, filename = 'students_with_remarks.csv'):
    rows = []
    found = False

    with open('students_with_remarks.csv', 'r', newline = '') as file:
        reader = csv.DictReader(file)
        header = reader.fieldnames

        for row in reader:
            if row['Name'].strip().lower() == Name.strip().lower():
                found = True
                continue

            rows.append(row)

    if not found:
        return

    with open(filename, 'w', newline = '') as file:
        writer = csv.DictWriter(file, fieldnames = header)
        writer.writeheader()
        writer.writerows(rows)
        print(f'Deleted{Name} (if present).')
        #return (f'Deleted{Name} (if present).')


#example
remove_student = delete_student('chris')
print(remove_student)
```

```python
import csv
import json

def csv_to_json(csv_filename="students_with_remarks.csv", json_filename="students.json"):
    data = []
    with open(csv_filename, "r") as f:
        reader = csv.DictReader(f)
        for row in reader:
            # Convert numeric fields to numbers
            # Try converting Score1, Score2, Average if they exist
            for key in ("Score1","Score2","Average"):
                if key in row and row[key] != "":
                    try:
                        # Average may be float; scores usually int
                        if key == "Average":
                            row[key] = float(row[key])
                        else:
                            row[key] = int(row[key])
                    except ValueError:
                        pass
            data.append(row)
    with open(json_filename, "w") as jf:
        json.dump(data, jf, indent=4)
    print(f"Wrote {len(data)} records to {json_filename}")

# Example usage:
csv_to_json()
```
