```python
score = np.array([70, 85, 90, 60, 78, 200])
```

```python
score
```

```python
np.sort(score)
```

```python
Q1 = np.percentile(score, 25)
Q1
```

```python
Q3 = np.percentile(score, 75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1 - 1.5 * 16.75


lower_bound
```

```python
higher_bound = Q3 + 1.5 * 16.75
higher_bound
```

```python
outlier = score[(score < lower_bound) | (score > higher_bound)]
outlier
```
