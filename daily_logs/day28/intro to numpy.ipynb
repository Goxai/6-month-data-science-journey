# Outliers detection

```python
nubs = np.array([10, 12, 14, 15, 18, 19, 20, 100])
nubs
```

```python
# IQR = Q3 - Q1

Q1 = np.percentile(nubs, 25)
Q1
```

```python
Q3 = np.percentile(nubs, 75)
Q3
```

```python
IQR = 19.25 - 13.5
IQR
```

```python
lower = Q1 - 1.5 * IQR
lower
```

```python
higher = Q3 + 1.5 * IQR
higher
```

```python
# detect outliers

outlier = nubs[(nubs < lower) | (nubs > higher)]
outlier
```

```python
# remove outliers

clean_outlier = nubs[(nubs >= lower) & (nubs <= higher)]
clean_outlier
```

# Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(nubs)
plt.title('original data')
plt.show()
```

```python
# outliers removed 

plt.plot(clean_outlier)
plt.title('no outliers')
plt.show()
```

```python
plt.hist(nubs, bins = 10)
plt.title('data distribution')
plt.show()
```

```python
plt.hist(clean_outlier, bins = 10)
plt.title('data distribution')
plt.show()
```

```python
plt.boxplot(nubs)
plt.title('orinal data')
plt.show()
```

```python
plt.boxplot(clean_outlier)
plt.title('outlier removed')
plt.show()
```

```python
plt.boxplot([nubs, clean_outlier])
plt.xticks([1,2], ['original', 'clean'])
plt.title('outlier removal comparison')
plt.show()
```
