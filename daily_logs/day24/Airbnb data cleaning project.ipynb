```python
'''
    checklist for this project
    
    drop duplicates
    delete redundant columns
    rename columns when needed
    clean individual columns
    rename the NaN values from the dataset
    check for some more transformations
'''
```

```python
# check for duplicates in the dataset

airbnb.duplicated().sum()
```

```python
# remove duplicates from the dataset

airbnb = airbnb.drop_duplicates()
airbnb
```

```python
airbnb.columns
```

```python
# rename column name e.g NAME

airbnb.rename(columns = {'NAME': 'name'}, inplace = True)
airbnb.head(3)
```

```python
airbnb.columns
```

```python
# drop the license column

airbnb = airbnb.drop(columns = ['license'])
airbnb
```

```python
# since index is available, i drop the id column

airbnb = airbnb.drop(columns = 'id')
airbnb
```

individual column cleaning

```python
# standardizing the name column

airbnb['name'].unique()
```

```python
# make the upper case in the name column to lower case

airbnb['name'] = airbnb['name'].str.lower()
airbnb
```

```python
# drop null values in the name column

#airbnb[airbnb['name'].isnull()]

airbnb = airbnb.dropna(subset = 'name')
airbnb
```

```python
# to check the if there is still null values in the name column

airbnb['name'].isnull().sum()
```

```python
airbnb.isnull().sum()
```

```python
# drop last review, house rules

airbnb.drop(columns = ['last review', 'house_rules', 'reviews per month'], inplace = True)
```

```python
airbnb.isnull().sum()
```

```python
# drop null values from the dataset

airbnb.dropna(inplace = True)
```

```python
airbnb.isnull().sum()
```

```python
# after removing null values, let look at individual columns 
# thenname column is fully cleaned 
# look at the host_identity_verified column


airbnb['host_identity_verified'].unique()
# the column is okay
```

```python
# the host name column is okay

airbnb['host name'].unique()
```

```python
# the construction year is in the float data type, we change it to date

airbnb['Construction year'].dtypes
```

```python
airbnb['Construction year'] = pd.to_datetime(airbnb['Construction year'])
airbnb['Construction year']
```

```python
airbnb['Construction year'].dtypes
```

```python
# from datetime to date data type

airbnb['Construction year'] = airbnb['Construction year'].dt.date
airbnb['Construction year']
```

```python
# the price column is in object data format as against int

airbnb['price'].dtypes
```

```python
# first remove the $ sign 

airbnb['price'] = airbnb['price'].str.lstrip('$')
airbnb['price']
```

```python
# before i continue, i reset the index column

airbnb = airbnb.reset_index()
airbnb
```

```python
# drop level_0 and index for clarity

airbnb = airbnb.drop(columns = ['level_0', 'index'])
airbnb
```

```python
# still on the price column

airbnb['price'] = airbnb['price'].str.replace(',', '')
airbnb['price']
```

```python
# convert the price data type to int

airbnb['price'] = airbnb['price'].astype(int)
airbnb['price']
```

```python
airbnb
```

```python
# we repeat the same we did for price column on the service fee column too

airbnb['service fee'].dtypes
```

```python
# remove the $ sign 

airbnb['service fee'] = airbnb['service fee'].str.lstrip('$')
airbnb['service fee']
```

```python
# to remove white spaces

airbnb['service fee'] = airbnb['service fee'].str.strip()
airbnb['service fee']
```

```python
# convert the 'service fee' to int data format

airbnb['service fee'] = airbnb['service fee'].astype(int)
airbnb['service fee']
```
