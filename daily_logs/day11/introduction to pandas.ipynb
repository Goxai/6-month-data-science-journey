```python
import pandas as pd
import csv

data = [
        ['name','age','score'],
        ['Alice',20,88],
        ['Bob',22,95],
        ['John',19,76],
        ['Mary',21,90]
]

with open('students_file.csv', 'w', newline = '') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

```python
with open('students_file.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
