```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\data-science-journey\jupyter notebooks files\students_data.csv')
df
```

```python
# will be conducting data cleaning on the dataset.

# first thing is to remove duplicate if there any from this dataset

df.duplicated()
```

```python
# since we have one duplicate, we remove it

df = df.drop_duplicates()
df
```

```python
# look at the entire dataset for null values

df.isnull().sum()
```

```python
# looking at the name column, there are irregularities on the alphabets like small letters with big letter
# have to standardize it so as to avoid future errors

df.loc[:, 'name'] = df['name'].str.lower().str.title()
df
```

```python
# look at the age column, fill the null values.
# looking close the age gap are, i filled the null values with the average age


df.loc[:, 'age'] = df['age'].fillna(df['age'].mean())
df
```

```python
#after removing all the null values from the age column, i convert to int

df.loc[:, 'age'] = df['age'].astype(int)
df
#df['age'].isnull()
```

```python
# standardized the gender column

df['gender'].unique()
```

```python
df.loc[:, 'gender'] = df['gender'].str.lower().str.title()
df
```

```python
# standardize the grade column

df['grade'].unique()
```

```python
df['grade'].dtypes
```

```python
# convert to string 

df.loc[:, 'grade'] = df['grade'].astype(str)
df
```

```python
# removing the 'th' from the numbers to standardize the grades

df.loc[:, 'grade'] = df['grade'].str.rstrip('th')
df
```

```python
# c0nvert the grade column from string to int

df.loc[:,'grade'] = df['grade'].astype(int)
df
```

```python
# null values in the math score column
df['math_score'].isnull().sum()
```

```python
# mean of the math score
df['math_score'].mean()
```

```python
# fill the null values in the math score column with the mean of the column

df.loc[:, 'math_score'] = df['math_score'].fillna(df['math_score'].mean())
df
```

```python
# the english score column

df['english_score']
```

```python
# convert the 'missing into a null value'
df.loc[:, 'english_score'] = pd.to_numeric(df['english_score'], errors = 'coerce')
df
```

```python
# fill the missing values in the english score column with the average score or mean

df.loc[:, 'english_score'] = df['english_score'].fillna(df['english_score'].mean()).round(1)
df
```

```python
df['english_score'].dtype
```

```python
# convert english score from string to int

df.loc[:, 'english_score'] = df['english_score'].astype(int)
df
```

```python
df['enrolled_date'].unique()
```

```python
# standardize the enrolled date to one fornat 
# dd-mm-yyyy format
dict_date = {
          '2022-06-10': '10-06-2022', 
           '06/12/2022': '06-12-2022',
            '2022/06/11': '11-06-2022'
}

# use a replace function 

df.loc[:, 'enrolled_date'] = df['enrolled_date'].replace(dict_date)
df
```

```python
# convert the enrolled date from string to datetime

df['enrolled_date'] = pd.to_datetime(df['enrolled_date'])
df
```

```python
df['enrolled_date'].dtype
```

```python
# because datetime is not what i need, i converted the datetime to just date

df['enrolled_date'] = df['enrolled_date'].dt.date
df
```

```python
df['enrolled_date'].dtype
```

```python
# standardize the remarks column


df.loc[:, 'remarks'] = df['remarks'].str.lower().str.title()
df
```
