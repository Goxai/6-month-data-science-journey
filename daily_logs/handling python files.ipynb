```python
import csv

students = []

with open('students.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        row['age'] = int(row['age'])
        row['score'] = int(row['score'])
        students.append(row)
```

```python
#to get the highest score

top_student = max(students, key = lambda x : x['score'])
print(top_student)
```

```python
#save summary into json


import json

summary = {
    'top_student': top_student['name'],
    'top_score': top_student['score']
}

with open('summary.json', 'w') as file:
    json.dump(summary, file, indent = 4)
```

```python
import csv

#find the top student without pandas
top_name = ''
top_avg = 0

with open('C:/Users/user/Desktop/data/students.csv', 'r') as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        math = int(row["Math"])
        eng = int(row["English"])
        
        avg = (math + eng) / 2
        
        if avg > top_avg:
            top_avg = avg
            top_name = row["Name"]
            
print('top studen:', top_name, 'with average:', top_avg)
```

```python
import csv

top_name = ""
top_avg = 0

with open('C:/Users/user/Desktop/data/students.csv', 'r') as f:
    reader = csv.DictReader(f)

    for row in reader:
        math = int(row["Math"])
        eng = int(row["English"])
        avg = (math + eng) / 2

        if avg > top_avg:
            top_avg = avg
            top_name = row["Name"]

print("Top student:", top_name, "with average of", top_avg)
```

```python
import csv

#find class statistics

scores = []
math_scores = []
eng_scores = []

with open('C:/Users/user/Desktop/data/students.csv', 'r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        math = int(row['Math'])
        eng = int(row['English'])

        scores.extend([math, eng])
        math_scores.append(math)
        eng_scores.append(eng)

print('class average:', sum(scores) / len(scores)),
print('math average:', sum(math_scores) / len(math_scores)),
print('eng average:', sum(eng_scores) / len(eng_scores))
```
