```python
def num_greater(numbs):
    num = 0                  # Step 1: counter
    for n in numbs:          # Step 2: loop through list
        if n > 10:           # Step 3: check condition
            num += 1         # Step 4: increase counter
    return num               # Step 5: return counter
```

```python
def sublist(numbers):
    odd_sub = 0
    for numb in numbers:
        sum_odd = 0
        for n in numb:
            if n % 2 != 0:
                sum_odd += n
        if sum_odd > odd_sub:
            odd_sub = sum_odd
    return odd_sub
```

```python
print(sublist(([[1,2,3], [4,5,6,7], [8,9]])))
```

```python
def largest_odd_sum(sublists):
    largest_sum = 0
    for sublist in sublists:
        sub_sum = 0
        for n in sublist:
            if n % 2 != 0:           # check if n is odd
                sub_sum += n          # add n to sub_sum
        if sub_sum > largest_sum:     # compare sub_sum with largest_sum
            largest_sum = sub_sum
    return largest_sum
```

```python
print(largest_odd_sum([[1,2,3], [4,5,6,7], [8,9]]))
```
