```python
def even_score_ranking(students):
    ranking = []
    for student in students:
        even_score = [score for score in student['scores'] if score % 2 == 0]
        if len(even_score) == 0:
            even_avg = 0
        else:
            even_avg = sum(even_score) / len(even_score)
        ranking.append((student['name'], even_avg))
    ranking.sort(key = lambda x : x[1], reverse = True)
    return ranking

def odd_score_ranking(students):
    ranking = []
    for student in students:
        odd_score = [score for score in student['scores'] if score % 2 != 0]
        if len(odd_score) == 0:
            odd_avg = 0
        else:
            odd_avg = sum(odd_score) / len(odd_score)
        ranking.append((student['name'], odd_avg))
    ranking.sort(key = lambda x : x[1], reverse = True)
    return ranking

def total_score_ranking(students):
    ranking = []
    for student in students:
        total_score = sum(student['scores'])
        ranking.append((student['name'], total_score))
    ranking.sort(key = lambda x : x[1], reverse = True)
    return ranking

def add_position(ranking):
    """ convert [('Alice' , 90), ('Bob', 80)] into
    [('list', 'Alice', 90), ('list', 'Bob', 80)]    
    """
    pos_suffix = {1: '1st', 2: '2nd', 3: '3rd'}
    
    final_list = []
    
    for i, (name, value) in enumerate(ranking, start = 1):
        pos = pos_suffix.get(i, f'{i}th')
        final_list.append((pos, name, value))
    return final_list

def class_average(students):
    #to get all the class average, even average, odd average, and total average.
    all_score = []
    all_even = []
    all_odd = []

    for student in students:
        for i in student['scores']:
            all_score.append(i)
            if i % 2 == 0:
                all_even.append(i)
            else:
                all_odd.append(i)

    return {
        "class_total_average": sum(all_score) / len(all_score),
        "class_even_average": sum(all_even) / len(all_even) if all_even else 0,
        "class_odd_average": sum(all_odd) / len(all_odd) if all_odd else 0        
    }

def full_student_analysis(students):
    #ranking
    even_rank = even_score_ranking(students)
    odd_rank = odd_score_ranking(students)
    total_rank = total_score_ranking(students)


    #add positions
    even_rank = add_position(even_rank)
    odd_rank = add_position(odd_rank)
    total_rank = add_position(total_rank)

    #class average
    class_stats = class_average(students)


    #build the final report

    return {
        "top_even": even_rank[0][1],
        "top_odd": odd_rank[0][1],
        "top_total": total_rank[0][1],
        "even_ranking": even_rank,
        "odd_ranking": odd_rank,
        "total_ranking": total_rank,
        "class_statistics": class_stats
    }
```

```python
def score_to_grade(students):
    for student in students:
        for score in student['scores']:
            if score >= 90:
                return "A"
            elif score >= 80:
                return "B"
            elif score >= 70:
                return "C"
            elif score >= 60:
                return "D"
            else:
                return "F"


def student_details(students):
    report = []
    for student in students:
        total_sum = sum(student['scores'])
        average = total_sum / len(student['scores'])
        percentage = average
        grade = score_to_grade(students)

        report.append({
            'name': student['name'],
            'total': total_sum,
            'average': average,
            'percentage': average,
            'grade': grade
        })
    return report


def full_student_analysis(students):
    #ranking
    even_rank = even_score_ranking(students)
    odd_rank = odd_score_ranking(students)
    total_rank = total_score_ranking(students)


    #add positions
    even_rank = add_position(even_rank)
    odd_rank = add_position(odd_rank)
    total_rank = add_position(total_rank)

    #class average
    class_stats = class_average(students)


    #students details
    details = student_details(students)


    #build the final report

    return {
        "top_even": even_rank[0][1],
        "top_odd": odd_rank[0][1],
        "top_total": total_rank[0][1],
        "even_ranking": even_rank,
        "odd_ranking": odd_rank,
        "total_ranking": total_rank,
        "class_statistics": class_stats,
        "student details" : details
    }
    
```

```python
def class_average(students):
    all_score = []
    all_even = []
    all_odd = []

    for student in students:
        for score in student['scores']:
            
            all_score.append(score)            #to get all score
            if score % 2 == 0:
                all_even.append(score)             #to get all even
            else:
                all_odd.append(score)              #to get all odd score

        #to calculate all avergaes 

        total_avg = sum(all_score) / len(all_score) if all_score else 0
        even_avg = sum(all_even) / len(all_even) if all_even else 0
        odd_avg = sum(all_odd) / len(all_odd) if all_odd else 0

    return {
        'class_total_avg': total_avg,
        'class_even_avg': even_avg,
        'class_odd_avg': odd_avg
    }
```

```python
def full_student_analysis(students):
    # 1️⃣ Get rankings
    even_rank = rank_even_score(students)
    odd_rank = rank_odd_score(students)
    total_rank = rank_total_score(students)

    # 2️⃣ Add positions (1st, 2nd, 3rd…)
    def add_positions(ranking):
        pos_suffix = {1: "1st", 2: "2nd", 3: "3rd"}
        final_list = []
        for i, (name, value) in enumerate(ranking, start=1):
            pos = pos_suffix.get(i, f"{i}th")
            final_list.append((pos, name, value))
        return final_list

    even_rank = add_positions(even_rank)
    odd_rank = add_positions(odd_rank)
    total_rank = add_positions(total_rank)

    # 3️⃣ Calculate class averages
    class_stats = class_average(students)

    # 4️⃣ Build final report
    report = {
        "top_even": even_rank[0][1],
        "top_odd": odd_rank[0][1],
        "top_total": total_rank[0][1],
        "even_ranking": even_rank,
        "odd_ranking": odd_rank,
        "total_ranking": total_rank,
        "class_statistics": class_stats
    }

    return report
```
