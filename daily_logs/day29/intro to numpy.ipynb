# Outliers detection

```python
nubs = np.array([10, 12, 14, 15, 18, 19, 20, 100])
nubs
```

```python
# IQR = Q3 - Q1

Q1 = np.percentile(nubs, 25)
Q1
```

```python
Q3 = np.percentile(nubs, 75)
Q3
```

```python
IQR = 19.25 - 13.5
IQR
```

```python
lower = Q1 - 1.5 * IQR
lower
```

```python
higher = Q3 + 1.5 * IQR
higher
```

```python
# detect outliers

outlier = nubs[(nubs < lower) | (nubs > higher)]
outlier
```

```python
# remove outliers

clean_outlier = nubs[(nubs >= lower) & (nubs <= higher)]
clean_outlier
```

```python
data = np.array([5, 7, 8, 9, 10, 12, 50])
data
```

```python
# detect outliers
# iqr = q3 - q1

q1 = np.percentile(data, 25)
q1
```

```python
q3 = np.percentile(data, 75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower = q1 - 1.5 * iqr
lower
```

```python
upper = q3 + 1.5 * iqr
upper
```

```python
# detect outlier

outlier = data[(data < lower) | (data > upper)]
outlier
```

```python
# remove outlier
clean_data = data[(data >= lower) & (data <= upper)]
clean_data
```

```python
# ploting histogram before outlier

plt.hist(data)
plt.title('data distribution')
plt.show()
```

```python
# plot histogram after removing outlier

plt.hist(clean_data)
plt.title('clean data distribution')
plt.show()
```

```python
plt.hist(data, label = 'original')
plt.hist(clean_data, label = 'cleaned')
plt.title('comparison')
plt.legend()
plt.show()
```

```python
# create a bosplot

plt.boxplot(data)
plt.title('boxplot of data')
plt.show()
```

```python
plt.boxplot(clean_data)
plt.title('boxplot of clean data')
plt.show()
```

```python
plt.boxplot([data, clean_data])
plt.xticks([1,2], ['original data', 'clean data'])
plt.title('before vs after outlier removal')
plt.show()
```

```python
plt.plot(data, marker = 'o', linestyle = '--')
```

```python
plt.plot(clean_data, marker = 'o', linestyle = '--')
```
