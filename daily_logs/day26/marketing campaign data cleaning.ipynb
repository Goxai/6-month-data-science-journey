```python
import pandas as pd 
import numpy as np

pd.set_option('display.max_columns', 12)

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\data-science-journey\Datasets\marketing_campaign_data_messy.csv')
df
```

```python
# column header cleaning to remove white spaces

df.columns
```

```python
# column header cleaning to remove white spaces

df.columns = df.columns.str.strip().str.lower()
```

```python
# to get the sum of duplicated data

df.duplicated().sum()
```

```python
# drop duplicates from the dataset

df.drop_duplicates(inplace = True)
df
```

```python
df.info()
```

```python
# to get the total null values in each columns

df.isnull().sum()
```

```python
# notice duplicate column called click
# removed the duplicated column

df = df.loc[:, ~df.columns.duplicated()]
```

```python
df.head()
```

# individual column cleaning

```python
# standardize the start date column from datetime to date 

df['start_date']
```

```python
# convert the start date first to date time to unify all dataset

df['start_date'] = pd.to_datetime(df['start_date'], format = 'mixed')
df['start_date']
```

```python
# convert from date time to date

df['start_date'].dt.date
```

```python
df
```

```python
# check for the unique values in the column

df['channel'].unique()
```

```python
# seeing the irregulaties, correct the spellings and reasign the names

df['channel'] = df['channel'].replace({'TikTok': 'tiktok', 'Facebook': 'facebook', 'Email': 'email', 'Instagram': 'instagram', 'Google Ads': 'google ads', 'E-mail': 'email',
       'Gogle': 'google ads', 'Tik_Tok': 'tiktok', 'Facebok': 'facebook', 'Insta_gram': 'instagra,'})
df['channel'].unique()
```

```python
# the spend column has the dollar sign and has to be removed

df['spend'] = df['spend'].str.lstrip('$')
df
```

```python
# convert the spend column to float from object

df['spend'] = df['spend'].astype(float)
df['spend']
```

```python
df
```

```python
# check the active column

df['active'].unique()
```

```python
# from the unique above, we can all see that they all mean the same thing
# we replace 0 = No, false = No, 1 = yes, Y = yes, true = yes

df['active'].replace({'Y': 'Yes', '0': 'No', 'No': 'No', 'True': 'Yes', 'Yes': 'Yes', '1': 'Yes', 'False': 'No'}, inplace = True)
df
```

```python
df['active'].unique()
```

```python
df.isnull().sum()
```

```python
# fill null values in the conversion column

df['conversions'] = df['conversions'].fillna(df['conversions'].mean()).round(0)
df['conversions'].isnull().sum()
```

```python
# drop the null values in the channel column 

df.dropna(subset = 'channel', inplace = True)
```

```python
df.isnull().sum()
```

```python
# handling outliers

Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)

IQR = Q3 - Q1
upper_limit = Q3 + (3 * IQR)

outlier = df['spend'] > upper_limit
print(df.loc[outlier, ['campaign_id', 'spend']].head(3))

df.loc[outlier, 'spend'] = upper_limit
print(df.loc[outlier, ['campaign_id', 'spend']].head(3))
```

```python
# extract seasons from the campaign name

df['season'] = df['campaign_name'].str.extract(r'Q\d_([^_]+)_')
df.head()
```

```python
df.to_csv('cleaned_marketing_campaign data.csv')
```
