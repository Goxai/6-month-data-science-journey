```python
# rename the 'adults', 'children', 'babies' column for clearer view of what the column means

df.rename({'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, axis = 1, inplace = True )
df.head()
```

```python
# drop the na values on the num children column

df.dropna(subset = ['num_children'], inplace = True)
```

```python
# drop the company column since it 94% 0f nan values, it is considered not important

df.drop(columns = 'company', inplace = True)
```

```python
# check the data types

df.dtypes
```

```python
# reasigned some columns to a proper data types

df = df.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int', 'agent': 'int' })
df.dtypes
```

```python
# want to group the lead time for easy readability

bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']
df['lead_time_binned'] = pd.cut(df['lead_time'], bins = bins, labels = labels)

df[['lead_time', 'lead_time_binned']]
```
