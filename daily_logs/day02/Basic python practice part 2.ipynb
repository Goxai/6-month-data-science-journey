```python
def total_package(num_hours, hourly_wage, tax_bracket):
    day_pay = num_hours * hourly_wage
    full_pay = day_pay * (1 - tax_bracket)
    return full_pay
```

```python
full_salary = total_package(40, 30, 0.3)
print(full_salary)
```

```python
full_salary = total_package(40, 24, 0.22)
print(full_salary)
```

```python
def get_expected_cost(beds, baths):
    cost_without_both = 80000
    cost_with_beds = 30000 * beds
    cost_with_baths = 10000 * baths
    value = cost_without_both + cost_with_beds + cost_with_baths
    return value
```

```python
def get_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    total_painted_space = sqft_walls + sqft_ceiling
    gallon_needed = total_painted_space / sqft_per_gallon
    cost = gallon_needed * cost_per_gallon
    return cost
```

```python
cost_of_painting = get_cost(432, 144, 400, 15)
print(cost_of_painting)
```

```python
cost_of_painting =get_cost(594, 288, 400, 15)
print(cost_of_painting)
```

```python
def get_costs(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    total_painted_space = sqft_walls + sqft_ceiling
    gallon_needed = round((total_painted_space / sqft_per_gallon), 3)
    cost = gallon_needed * cost_per_gallon
    return cost
```

```python
cost_of_painting = get_costs(594, 288, 400, 15)
print(cost_of_painting)
```

```python
def get_grade(score):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    return grade
```

```python
def largest_odd(numbers):
    number = []                     # Step 1: collect odd numbers
    for n in numbers:               # Step 2: loop through original list
        if n % 2 != 0:
            number.append(n)
            #return number
    
    if not number:                  # Step 3: check if odd list is empty
        return None
    
    largest = number[0]             # Step 4: assume first odd number is largest
    for n in number:                # Step 5: loop through odd numbers
        if n > largest:
            largest = n
    
    return largest                  # Step 6: return the largest odd number
```

```python
print(largest_odd([4, 7, 2, 9, 12]))
```

```python
def largest_odd_sum(sublist):
    largest_sum = 0
    for sub in sublist:
        sub_sum = 0
        for n in sub:
            if n % 2 != 0:
                sub_sum += n
                
        if sub_sum > largest_sum:
            largest_sum = sub_sum
    return largest_sum
            
```

```python
print(largest_odd_sum([[1,2,3], [4,5,6,7], [8,9]]))
```
