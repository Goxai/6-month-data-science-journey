```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([5, 7, 8, 9, 10, 12, 50])
data
```

```python
q1 = np.percentile(data, 25)
q1
```

```python
q3 = np.percentile(data, 75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower = q1 - 1.5 * iqr
lower
```

```python
upper = q3 + 1.5 * iqr
upper
```

```python
outlier = data[(data < lower) | (data > upper)]
outlier
```

```python
clean_data = data[(data >= lower) & (data <= upper)]
clean_data
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10, 4))

ax[0, 0].hist(data)
ax[0, 0].set_title('original data')

ax[0, 1].boxplot(data)
ax[0, 1].set_title('original data boxplot')

ax[1, 0].hist(clean_data)
ax[1, 0].set_title('clean data')

ax[1, 1].boxplot(clean_data)
ax[1, 1].set_title('clean data boxplot')


plt.show()
```

```python
iqr = q3 - q1
iqr
```

```python
lower = q1 - 1.5 * iqr
lower
```

```python
upper = q3 + 1.5 * iqr
upper
```

```python
outlier = sale[(sale < lower) | (sale > upper)]
outlier
```

```python
clean_sale = sale[(sale >= lower) & (sale <= upper)]
clean_sale
```

```python
# lin plot of the original data

plt.figure(figsize = (10,5))
plt.plot(sale, marker = 'o', linestyle = '--', color = 'green', linewidth = 3)
plt.xlabel('index')
plt.ylabel('values')
plt.title('original sale data')
plt.grid(True)
plt.show()
```

```python
# line plot of the clean data

plt.figure(figsize = (10,5))
plt.plot(clean_sale, marker = 'o', linestyle = '--', color = 'green', linewidth = 3)
plt.xlabel('index')
plt.ylabel('values')
plt.title('original sale data')
plt.grid(True)
plt.show()
```

```python
# subplot for the sale data

fig, ax = plt.subplots(2, 2, figsize = (10, 7))

ax[0, 0].hist(sale)
ax[0, 0].set_title('sale distribution')

ax[0, 1].boxplot(sale)
ax[0, 1].set_title('sale boxplot')

ax[1, 0].hist(clean_sale)
ax[1, 0].set_title('clean sale distribution')

ax[1, 1].boxplot(clean_sale)
ax[1, 1].set_title('clean sale boxplot')

plt.show()
```
